## Problem 2:
Define a function to add new row data to a CSV file.

```python
import pandas

def readCSV(filename):
    df=pandas.read_csv(filename)
    return df

def addrow(filename,rowdata):
    df = readcsv(filename)
    df.to_csv(filename,index=False)
    df.loc[len(df)] = rowdata
    return df

rowdata=['010101','statename',123,340,345,23456,34678,2345,45677,34556]
addrow('DataFiles/income.csv',rowdata)
```

## Problem 3:
Define a function to update/modify a row in a csv file.

```python
def updaterow(filename,rowindex,rowdata):
    df=readcsv(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)
    return df

rowdata=['020202','state1',1,2,3,4,5,6,7,8]
updaterow('DataFiles/income.csv',6,rowdata)
```

## Problem 4:
Define a function to delete a row from a csv file.

```python
def deleterow(filename,rowindex):
    df=readcsv(filename)
    df=df.drop(rowindex)       # to delete a row
    df.to_csv(filename,index=False)
    return df

deleterow('Datafiles/income.csv',6)
```

## Problem 5:
