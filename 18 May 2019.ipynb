## Problem 1:
Plotting 2D graphs using matplotlib.

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

plt.plot([1,2,3,4],[1,8,27,64])
plt.xlabel('time')
plt.ylabel('distance')
plt.show()
```

```python
import numpy as np

t = np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('sine wave')
plt.plot(t,v)
plt.grid(True)
plt.savefig('figures/time-voltage.png')
plt.show()
```

## Problem 2:
Plot a line graph on income versus year for the state of california.

```python
import pandas as pd

df=pd.read_csv('Datafiles/income.csv')
y=df.values[4]
x=df.columns

```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income
plt.xlabel('year')
plt.ylabel('income')
plt.title('California state')
plt.plot(years,income)
plt.savefig('Figures/line-graph-income.png')
plt.grid(True)
plt.show()
```

```python
width=1/2
plt.xlabel('year')
plt.ylabel('income')
plt.title('Income of California')
plt.bar(years,income,width,color="blue")
plt.savefig('Figures/bar-graph-income.png')
plt.show()
```

## Problem 4:
Generate a scatter plot of 50 random numbers.

```python
n=50
x = np.random.rand(n)
y = np.random.rand(n)
colorlist = np.random.rand(n)
arealist = np.pi * (15 * np.random.rand(50))**2
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)
plt.savefig('Figures/scatter-plot-png')
plt.show()
```
